```python
# Select all numerical data, except column Attrition as this column are variable Y/target variable

X_num = employee_df[['Age', 'DailyRate','DistanceFromHome','Education','EnvironmentSatisfaction','HourlyRate','JobInvolvement','JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate','NumCompaniesWorked','OverTime','PercentSalaryHike','PerformanceRating','RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]
X_num
```

```python
# Combine both category and numerical 
X_all = pd.concat([X_cat, X_num], axis = 1)
X_all
```

# Data Scaling

```python
# import the library
# MinMaxScaler is used because standardscaler cannot be used. Reason being that the dataset contains outliers

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Create variable X 
X = scaler.fit_transform(X_all)
X

```

```python
# Create variable Y
y = employee_df['Attrition']
y
```

# Split Data

```python
# Split the data with the test size 25%
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
# Check the size/shape
X_train.shape
```

```python
# Check the size/shape
X_test.shape
```

# Data modelling

```python
# Import model libraries

from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
# Create the model variable 
svc = SVC(kernel = 'linear')
lr = LogisticRegression(solver = 'saga', multi_class = 'auto', max_iter = 550, fit_intercept = True)
knn = KNeighborsClassifier(n_neighbors = 3)
nb = GaussianNB()
dt = DecisionTreeClassifier(max_depth = 80)

models = ['svc','lr','knn','nb','dt']
```

```python
for i in range(len(models)):
    print(models[i]+"\n")
    eval(models[i]).fit(X_train, y_train)
    y_pred = eval(models[i]).predict(X_test)
    print("Accuracy: ", eval(models[i]).score(X_test, y_test))
    print("\n")
    print("Confusion Matrix: ")
    print(confusion_matrix(y_test, y_pred))
    print("\n")
    print("Classification Report: ")
    print(metrics.classification_report(y_test, y_pred))
    print()
    print("======================================================")
   
```


# Apply Principal Component Analysis(PCA)

X and y variable already being scaled before

```python
# Check the Xtrain shape/size
X_train.shape
```

```python
# Import PCA library
from sklearn.decomposition import PCA
```

```python
# Create PCA model


#Reduce dimensionality from 50 to 25
pca_model = PCA(25, svd_solver = 'randomized').fit(X) 

Xtrain_pca = pca_model.transform(X_train)
Xtest_pca = pca_model.transform(X_test)
```

```python
# Check the feature size after being reduce. Make sure the size are 25
Xtrain_pca.shape
```

# Data Modelling with PCA

```python
for i in range(len(models)):
    eval(models[i]).fit(Xtrain_pca, y_train)
    y_pred_pca = eval(models[i]).predict(Xtest_pca)
    print(models[i] + " With PCA : \n")
    print("Accuracy : \n", eval(models[i]).score(Xtest_pca, y_test))
    print("\nConfusion Matrix : \n", confusion_matrix(y_test, y_pred_pca))
    print("\nClassification Report : \n", metrics.classification_report(y_test, y_pred_pca))
    print("====================================================")
```

# TensorFlow

```python
# import library
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu', input_shape = (50,)))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam', loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 50)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```
